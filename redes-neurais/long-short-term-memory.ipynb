```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import EarlyStopping

# Definir parâmetros da sequência e número de features
sequencia = 10      # Número de passos de tempo (timesteps)
features = 5        # Número de variáveis por passo de tempo (features)

# Gerar dados de treino e teste fictícios
X_treino = np.random.rand(100, sequencia, features)  # 100 amostras de treino
y_treino = np.random.rand(100, 1)                    # 100 saídas de treino

X_teste = np.random.rand(20, sequencia, features)    # 20 amostras de teste
y_teste = np.random.rand(20, 1)                      # 20 saídas de teste

# Gerar novos dados para previsão
novos_dados = np.random.rand(5, sequencia, features) # 5 novas amostras

# Criar modelo
modelo = Sequential()
modelo.add(LSTM(50, return_sequences=False, input_shape=(sequencia, features)))
modelo.add(Dense(1))
modelo.compile(optimizer='adam', loss='mse')

# Callback para parar o treinamento se a validação não melhorar
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Treinar modelo
modelo.fit(X_treino, y_treino, epochs=100, 
           validation_data=(X_teste, y_teste),
           callbacks=[early_stop],
           verbose=1)

# Fazer previsão
previsao = modelo.predict(novos_dados)

# Mostrar previsão
print("Previsões:")
print(previsao)
```
