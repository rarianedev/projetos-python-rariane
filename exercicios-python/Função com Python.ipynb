```python
alunos = []

def cadastrar_aluno():
    nome = input("\nNome do aluno: ")
    notas = []
    for i in range(4):
        nota = float(input(f"Nota {i + 1}: "))
        notas.append(nota)
    media = sum(notas) / len(notas)
    print("\nA média do aluno é: ", media)
    aluno = {"nome": nome, "notas": notas, "media": media}
    alunos.append(aluno)

def listar_alunos():
    if not alunos:
        print("Nenhum aluno.")
        return
    for aluno in alunos:
        print(f"{aluno['nome']}, notas: {aluno['notas']}, média: {aluno['media']:.2f}")

def buscar_aluno(nome):
    for aluno in alunos:
        if aluno['nome'].lower() == nome.lower():
            print(f"{aluno['nome']}, média: {aluno['media']:.2f}")
            return
    print("Aluno não encontrado.")

def exibir_aprovados():
    for aluno in alunos:
        if aluno['media'] >= 6:
            print(f"{aluno['nome']} aprovado com média {aluno['media']:.2f}")

def excluir_cadastro(nome):
    global alunos
    alunos = [aluno for aluno in alunos if aluno['nome'].lower() != nome.lower()]
    print("\nCadastro excluído.")

def encerrar_programa():
    print("\nFim do programa.")

def menu():
    while True:
        print("\n----- MENU -----")
        print("1. Cadastrar aluno")
        print("2. Listar alunos")
        print("3. Buscar aluno")
        print("4. Exibir aprovados")
        print("5. Excluir cadastro")
        print("6. Encerrar programa")
        opcao = input("\nEscolha uma opção: ")

        if opcao == '1':
            cadastrar_aluno()
        elif opcao == '2':
            listar_alunos()
        elif opcao == '3':
            nome = input("\nDigite o nome do aluno para buscar: ")
            buscar_aluno(nome)
        elif opcao == '4':
            exibir_aprovados()
        elif opcao == '5':
            nome = input("\nDigite o nome do aluno para excluir: ")
            excluir_cadastro(nome)
        elif opcao == '6':
            encerrar_programa()
            break
        else:
            print("\nOpção inválida. Tente novamente.\n")
menu()
```
