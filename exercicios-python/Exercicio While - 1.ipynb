```python
#1  Imprima os números de 1 a 10 usando while. 
i = 1
while i <= 10:
    print(i)
    i += 1
```

```python
#2 Imprima os números pares de 2 a 20 usando while. 
i = 2
while i <= 20:
    if i%2==0:
        print(i)
    i+=1
```

```python
#3 Some os números de 1 a 100 usando while.
i = 1
soma = 0

while i <= 100:
        soma+=i
        i+=1
print(soma)
```

```python
#4. Conte regressivamente de 10 até 1 usando while. 
i = 10

while i>1:
    print(i)
    i=i-1
print(i)
```

```python
#5. Peça ao usuário uma senha até que ele digite a correta. 

senha_correta = '123'

while True:
    senha = input('Informe a senha: ')
    if senha == senha_correta:
        print('Senha correta')
        break
    else: 
        print('Senha incorreta, tente novamente')
```

```python
#6. Peça um número positivo. Enquanto o número for negativo, repita a pergunta. 
while True:
    numero = int(input('Informe um numero: '))
    if numero > 0:
        print('Numero positivo')
        break
    elif numero == 0:
        print('sem polaridade')
    else:
        print('Numero negativo')
```

```python
#7. Calcule a média de notas inseridas até que o usuário digite -1. 
soma = 0.0
cont = 0

while True:
    n = float(input('Informe a nota (ou -1 para sair): '))
    if n == -1:
        break
    soma += n
    cont += 1

if cont > 0:
    media = soma / cont
    print(f'Média: {media:.2f}')
else:
    print('Nenhuma nota informada.')
```
