```python
#Função

def cadastrar_usuario():
    nome = input('Nome: ')
    cpf = input('CPF: ')
    idade = int(input('Idade: '))
    print('\nUsuario cadastrado com sucesso! \nNome: {} \nCPF: {} \nIdade: {}'.format(nome,cpf,idade))

cadastrar_usuario()
```

```python
def media_de_notas(n1,n2,n3,n4):
    return (n1+n2+n3+n4)/4

n1 = float(input('Informe a sua 1° nota: '))
n2 = float(input('Informe a sua 2° nota: '))
n3 = float(input('Informe a sua 3° nota: '))
n4 = float(input('Informe a sua 4° nota: '))

media = media_de_notas(n1,n2,n3,n4)
print('Sua media é: {}'.format(media))

if media<6:
    print('Não passou -Recuperação')
elif media>=6:
    print('Passou direto')
```
