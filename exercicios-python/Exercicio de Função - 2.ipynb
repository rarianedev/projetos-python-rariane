```python
def eh_alcoolica(bebida):
    bebida = bebida.upper()
    if 'BEB' in bebida:
        return True
    else:
        return False

produtos = ['beb46275','TFA23965',]
for produto in produtos:
    if eh_alcoolica(produto):
        print('Enviar o produto {} para a area de bebidas alcoolicas.'.format(produto))     
```
