```python
#definindo nosso classificador
detector_carros = cv2.CascadeClassifier('/content/drive/MyDrive/VISAO/VISAO/Cascades-20250722T022546Z-1-001/Cascades/cars.xml')
#definindo a imagem
imagem = cv2.imread('/content/drive/MyDrive/VISAO/VISAO/Images-20250722T160930Z-1-001/Images/car.jpg')
#altera o tamanho da imagem e informa seu novo tamanho
imagem = cv2.resize(imagem, (800, 600))
#coloca a imagem em escala de cinza
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

#realiza as detecções
deteccoes_carros = detector_carros.detectMultiScale(imagem_cinza, scaleFactor=1.02, minNeighbors=2, minSize=(30, 30))

for x, y, w, h in deteccoes_carros:
  print(w,h)
  cv2.rectangle(imagem, (x, y), (x + w, y + h), (75, 255, 0), 2)

quantidade_carros = len(deteccoes_carros)
print(f'Quantidade de carros na imagem: {quantidade_carros}')

cv2_imshow(imagem)
```
