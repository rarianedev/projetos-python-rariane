```python
#definindo nosso classificador
detector_rosto = cv2.CascadeClassifier('/content/drive/MyDrive/VISAO/VISAO/Cascades-20250722T022546Z-1-001/Cascades/haarcascade_frontalface_default.xml')
#definindo a imagem
imagem = cv2.imread('/content/drive/MyDrive/VISAO/VISAO/Images-20250722T160930Z-1-001/Images/people5.jpg')
#altera o tamanho da imagem e informa seu novo tamanho
imagem = cv2.resize(imagem, (800, 600))
#coloca a imagem em escala de cinza
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

#realiza as detecções
deteccoes_rostos = detector_rosto.detectMultiScale(imagem_cinza, scaleFactor=1.02, minNeighbors=2, minSize=(30, 30))

for x, y, w, h in deteccoes_rostos:
  print(w,h)
  cv2.rectangle(imagem, (x, y), (x + w, y + h), (75, 255, 0), 2)

quantidade_rostos = len(deteccoes_rostos)
print(f'Quantidade de rostos na imagem: {quantidade_rostos}')

cv2_imshow(imagem)
```
