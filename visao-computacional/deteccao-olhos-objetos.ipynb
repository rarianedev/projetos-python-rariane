Detecçao de Olhos com Haar Cascade

```python
import cv2
```

```python
from google.colab.patches import cv2_imshow #cv2_imsow para visualização das imagens
from google.colab import drive #montando o drive google para acessar os arquivos
drive.mount('/content/drive')
```

```python
detector_olhos = cv2.CascadeClassifier('/content/drive/MyDrive/VISAO/VISAO/Cascades-20250722T022546Z-1-001/Cascades/haarcascade_eye.xml')
```

```python
imagem = cv2.imread('/content/drive/MyDrive/VISAO/VISAO/Images-20250722T160930Z-1-001/Images/people2.jpg')
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

#definindo nosso classificador
detector_rosto = cv2.CascadeClassifier('/content/drive/MyDrive/VISAO/VISAO/Cascades-20250722T022546Z-1-001/Cascades/haarcascade_frontalface_default.xml')
deteccoes = detector_rosto.detectMultiScale(imagem_cinza, scaleFactor=1.1, minNeighbors=2, minSize=(30,30))
#print(deteccoes)

for (x, y, w, h) in deteccoes: #Incia um loop para percorrer todas as posições de faces
  cv2.rectangle(imagem, (x, y), (x + w, y + h), (0, 255, 0), 2) #em cada face detectada vamos desenhar um retangulo para identificação
  deteccoes_olhos = detector_olhos.detectMultiScale(imagem_cinza, scaleFactor=1.1, minNeighbors=3, minSize=(20,20), maxSize=(55,55))
  #for (x2, y2, w2, h2) in deteccoes_olhos:
    #cv2.rectangle(imagem, (x2, y2), (x2 + w2, y2 + h2), (255, 0, 0), 2)
  for (x, y, w, h) in deteccoes_olhos:
    cv2.rectangle(imagem, (x, y), (x + w, y + h), (255, 0, 0), 2)
cv2_imshow(imagem)
```

```python
detector_objetos = cv2.CascadeClassifier('/content/drive/MyDrive/VISAO/VISAO/Cascades-20250722T022546Z-1-001/Cascades/clocks.xml')
imagem = cv2.imread('/content/drive/MyDrive/VISAO/VISAO/Images-20250722T160930Z-1-001/Images/clock.jpg')
#imagem = cv2.resize(imagem, (1920, 1800))
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

#realiza as detecções
deteccoes_objetos = detector_objetos.detectMultiScale(imagem_cinza, scaleFactor=1.03, minNeighbors=1, minSize=(25, 25))

for x, y, w, h in deteccoes_objetos:
  print(w,h)
  cv2.rectangle(imagem, (x, y), (x + w, y + h), (75, 255, 0), 2)

quantidade_carros = len(deteccoes_objetos)
print(f'Quantidade de objetos na imagem: {quantidade_carros}')

cv2_imshow(imagem)
```

```python
imagem = cv2.imread('/content/drive/MyDrive/VISAO/VISAO/Images-20250722T160930Z-1-001/Images/people2.jpg')
cv2_imshow(imagem)
```

```python
detector_face = dlib.get_frontal_face_detector()
```

```python
deteccoes_face = detector_face(imagem, 4)
```

```python
deteccoes_face, len(deteccoes_face)
```

```python
for face in deteccoes_face:
  l, t, r, b = face.left(), face.top(), face.right(), face.bottom()
  cv2.rectangle(imagem, (l, t), (r, b), (255, 0, 70), 2)
cv2_imshow(imagem)
```
